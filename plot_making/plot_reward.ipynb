# Plots for the presentation

We need to plot the rewards regarding the time steps, for the best, worst and average case.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("best_worst_mean_rewards.csv")
df.head()
```

Exclude "Category" as a column.

```python
best_row = df[df['Category'] == 'Best'].iloc[0, 1:]  
worst_row = df[df['Category'] == 'Worst'].iloc[0, 1:]  
average_row = df[df['Category'] == 'Mean'].iloc[0, 1:]
```

We now plot for the custom environment.

```python
# Plot the values
plt.plot(average_row.values, color='black', linewidth=2.5, label='Average Value')
plt.plot(best_row.values, color='green', linewidth=2.5, label='Best Simulation')
plt.plot(worst_row.values, color='red', linewidth=2.5, label='Worst Simulation')

# Labels
plt.ylabel('Reward')
plt.xlabel('Timesteps')
plt.title('Performance Evaluation for Custom Environment')
plt.legend()
plt.grid(True)
plt.savefig("custom_rewards", dpi=300, bbox_inches='tight') # export

plt.show()
```
